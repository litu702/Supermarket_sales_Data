# Q1. Which branch has the more income peecentage

```python
results=df.groupby(['Branch']).sum()
results
```

```python
import matplotlib.pyplot as plt

Branches = [branch for branch, df in df.groupby(['Branch'])]

plt.bar(Branches,results['gross income'])
plt.xticks(Branches)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Branches')
plt.show()
```

```python
plot = results.plot.pie(y='gross income', figsize=(7, 7))
```

```python
plot = results.plot.pie(y='gross income', title="Gross income Percentage for each Branch", legend=False, \
                   autopct='%1.1f%%', explode=(0, 0, 0.1), \
                   shadow=True, startangle=0)
```

# Q2. Which type of customer has the most expenditure

```python
cust_types=df.groupby(['Customer type']).sum()
cust_types
```

```python
customers = [customer for customer, df in df.groupby(['Customer type'])]

plt.bar(customers,cust_types['Total'])
plt.xticks(customers)
plt.ylabel('Expenditure in USD ($)')
plt.xlabel('Type of Customer')
plt.show()
```

```python
plot = cust_types.plot.pie(y='Total', title="Expenditure of each customer types", legend=False, \
                   autopct='%1.1f%%', explode=(0, 0.1), \
                   shadow=True, startangle=0)
```

# Q3. Which product line has the most contribution to the income?

```python
product_group=df.groupby(['Product line']).sum()
product_group
```

```python
products=[product for product, df in df.groupby(['Product line'])]
plt.bar(products,product_group['gross income'])
plt.xticks(products, rotation='vertical', size=8)
plt.xlabel("Product Line")
plt.ylabel("Gross Income in ($)")
plt.show()
```

```python
plot = product_group.plot.pie(y='gross income', title="Gross income for each product line", legend=False, \
                   autopct='%1.1f%%', explode=(0,0,0.1,0,0,0),\
                   shadow=True, startangle=0)
#plt.legend(loc='upper right')
```
