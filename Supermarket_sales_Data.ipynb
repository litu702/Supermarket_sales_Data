# Q1. Which branch has the more income peecentage

# By means of gross income

```python
results=df.groupby(['Branch']).sum()
results
```

```python
import matplotlib.pyplot as plt

Branches = [branch for branch, df in df.groupby(['Branch'])]

plt.bar(Branches,results['gross income'])


plt.xticks(Branches)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Branches')
plt.show()
```

```python
plot = results.plot.pie(y='gross income', figsize=(7, 7))
```

```python
plot = results.plot.pie(y='gross income', title="Gross income Percentage for each Branch", legend=False, \
                   autopct='%1.1f%%', explode=(0, 0, 0.1), \
                   shadow=True, startangle=0)
```

```python
Branches = [branch for branch, df in df.groupby(['Branch'])]

plt.bar(Branches,results['gross margin percentage'])


plt.xticks(Branches)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Branches')
plt.show()
```

# Which day of a particular branch has the maximum gross income

```python
df['Day']=pd.to_datetime(df['Date']).dt.day
df['Count']=1
df.head()
```

```python
#groupby_day=df.groupby(['Month', 'Day']).sum()
#groupby_day=df.groupby(['Branch','Day']).count()
groupby_day=df.groupby(['Branch','Day']).sum().reset_index()
groupby_day
```

```python
groupby_day['gross_income'].max()
```

```python
results=groupby_day[['Branch','Day','gross_income']]
results
```

```python
df.rename(columns={'gross income': 'gross_income'}, inplace=True)
```

```python
branch_A = results.loc[results.Branch=="A"]
branch_B = results.loc[results.Branch=="B"]
branch_C = results.loc[results.Branch=="C"]
#branch_A
#Making two separate list for branch A to draw the bar chart
#x=branch_A['Day']
#y=branch_A['gross income']
```

```python
plt.figure(figsize=(15,8))
plt.title('Daily Gross Income for Branches', fontdict={'fontweight':'bold', 'fontsize': 18})
plt.plot(branch_A.Day, branch_A.gross_income, 'b.-', label='Branch A')
plt.plot(branch_B.Day, branch_B.gross_income, 'r.-', label='Branch B')
plt.plot(branch_C.Day, branch_C.gross_income, 'g.-', label='Branch C')
plt.xticks(branch_A.Day)
plt.xlabel('Day')
plt.ylabel('gross income')
plt.legend()


plt.show()
```
